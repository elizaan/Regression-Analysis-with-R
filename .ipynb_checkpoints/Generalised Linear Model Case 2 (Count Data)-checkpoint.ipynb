# We want to know if Tree replantation activities have affected the count of Tigers, we fit GLM model. The response variable is count of the Tigers, so we use family = poisson.

```r
forest <- read.csv("forest.csv", header = T)
Replant<- glm(Tiger~ForestTreatment, family="poisson",data=forest)
summary(Replant)
```

# Histogram of the Tiger Count

```r
hist(forest$Tiger)
```

# Plotting Forest Replantation/ Regeneration vs Tiger Count

```r
plot(Tiger~ForestTreatment, data = forest, xlab = "Forest Revegenation", ylab =
"Tiger Count",pch=16)
```

# Plotting the residuals

```r
plot(Replant,which=1)
```

# Also, running Anova to learn about the interpretation again

```r
anova(Replant,test="Chisq")
```

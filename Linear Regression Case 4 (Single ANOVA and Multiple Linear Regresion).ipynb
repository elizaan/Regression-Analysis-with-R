# Single factor ANOVA Linear Model for yield and soil values from yields data

```r
yields = read.csv("yields.csv", header = TRUE)
yields$soil <- factor(yields$soil)
yields.aov <- aov(yield ~ soil, data = yields)
summary(yields.aov)
```

# Boxplot of Single factor ANOVA Linear Model for yield and soil values from yields data

```r
boxplot(yield~soil, data=yields, ylab = "Yield", xlab = "Soil")
```

# Identifying which groups are different in Single Factor ANOVA Linear Model for yield and soil values from yields data

```r
TukeyHSD(yields.aov)
```

```r
yields$soil <- factor(yields$soil)
yields$mineral.percentage. <- factor(yields$mineral.percentage.)
yields.aov <- aov(yield ~ soil*mineral.percentage., data = yields)
summary(yields.aov)
```

# Boxplot of Single factor ANOVA Linear Model for yield and mineral values from yields data

```r
boxplot(yield~mineral.percentage., data=yields, ylab = "Yield", xlab = "Mineral (%)")
```

# Boxplot of Multiple Linear Model for yield vs soil and mineral values from yields data

```r
boxplot(yield~ soil*mineral.percentage., data=yields, ylab = "Yield", xlab = "Soil x Mineral percentage")
```

# Identifying which groups are different in Multiple Linear Model for yield vs soil and mineral values from yields data

```r
TukeyHSD(yields.aov)
```
